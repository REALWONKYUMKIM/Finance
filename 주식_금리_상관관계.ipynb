```python
sp500_data1 = pd.read_csv("./datasets/sp500_data1.csv")
sp500_data2 = pd.read_csv("./datasets/sp500_data2.csv")
sp500_data = pd.concat([sp500_data1, sp500_data2], axis=1)
sp500_data.head()
```
