```python
df_sp = yf.download("^GSPC", start="2008-01-03", end="2017-12-29")
df_sp.head()
```

```python
df_kospi200 = yf.download("^KS200", start="2008-01-03", end="2017-12-29")
df_kospi200.head()
```

```python
plt.figure(figsize=(9, 6))
plt.title('SP vs KOSPI200 close')
plt.plot(df_sp.index, (df_sp['Adj Close'] / df_sp['Adj Close'][0])*100, label='S&P500 Indexed')
plt.plot(df_kospi200.index, (df_kospi200['Adj Close'] / df_kospi200['Adj Close'][0])*100 , label='KOSPI200 Indexed')
plt.legend()
plt.show()
```
